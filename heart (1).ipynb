```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV  # <-- Add this import statement
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score  # <-- Add this import statement
```

```python
#connecting kaggle dataset
```

```python
heart_data = pd.read_csv('/heart_disease_data.csv')
```

```python
# checking the distribution of Target Variable
#0 means helthy and 1 means heart has problem
heart_data['target'].value_counts()
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
scaler = StandardScaler()  #  Initialize the StandardScaler
X = scaler.fit_transform(X)  # Apply scaling to the features
```

```python
print(X)
print(Y)
```

```python
#train test split= 20% test and now train on the 80%data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Tuning the logistic regression model using GridSearchCV
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}  # <-- Change 3: Grid of hyperparameters
model = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='accuracy')  # <-- Change 4: Use GridSearchCV
model.fit(X_train, Y_train)
```

```python
print('Best Parameters:', model.best_params_)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on Training data :', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
precision = precision_score(Y_test, X_test_prediction)
recall = recall_score(Y_test, X_test_prediction)
f1 = f1_score(Y_test, X_test_prediction)
roc_auc = roc_auc_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on Test data :', test_data_accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1)
print('ROC AUC Score:', roc_auc)
```

```python
# Define the feature names based on your original dataset
feature_names = heart_data.columns[:-1]  # All columns except 'target'

# Example prediction input
input_data = (62, 0, 0, 140, 268, 0, 0, 160, 0, 3.6, 0, 2, 2)

# Convert the input data to a DataFrame with the correct feature names
input_data_df = pd.DataFrame([input_data], columns=feature_names)

# Scale the input data
input_data_scaled = scaler.transform(input_data_df)  # Scale the input data

# Make a prediction
prediction = model.predict(input_data_scaled)  # Use the scaled input data for prediction
print(prediction)

if (prediction[0] == 0):
    print('The Person does not have a Heart Disease')
else:
    print('The Person has Heart Disease')
```
